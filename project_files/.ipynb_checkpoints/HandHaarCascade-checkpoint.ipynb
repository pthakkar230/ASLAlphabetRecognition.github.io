```python
import sys
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("asl_alphabet.png")
image_bgr = image[:,:,::-1]
image_rgb = image[:,:,::-1]
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

handCascade = cv2.CascadeClassifier('hand.xml')

hands = handCascade.detectMultiScale(image_gray, scaleFactor=1.03, minNeighbors=1, minSize=(40,40))

print(len(hands))
```

```python
print("[INFO] Found {0} Hands!".format(len(hands)))

print(hands)

image_bgr = np.ascontiguousarray(image_rgb, dtype=np.uint8)

for (x, y, w, h) in hands:
    x = int(x)
    y = int(y)
    w = int(w)
    h = int(h)
    cv2.rectangle(image_bgr, (x,y), (x + w, y + h), (0, 255, 0), 2)
    
status = cv2.imwrite('hands_detected.jpg', image_bgr)

fig = plt.figure()
fig.set_size_inches(20,15)

plt.imshow(image_bgr, aspect="auto")
```
